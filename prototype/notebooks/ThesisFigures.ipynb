```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use('seaborn-paper')

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
# set matplotlib font to Times
from matplotlib import rc
rc('font',**{'family':'serif','serif':['Times']})
rc('text', usetex=True)
```

## Asymmetric mean squared error

```python
f = plt.figure()

x = np.linspace(-5, 5)
plt.plot(x, (x**2) * ((np.sign(x)) ** 2), 'k', label=r"$\alpha=0$")
plt.plot(x, (x**2) * ((np.sign(x) + 0.2) ** 2), 'k:', label=r"$\alpha=0.2$")
plt.plot(x, (x**2) * ((np.sign(x) + 0.4) ** 2), 'k-.', label=r"$\alpha=0.4$")
plt.plot(x, (x**2) * ((np.sign(x) + 0.6) ** 2), 'k--', label=r"$\alpha=0.6$")
plt.xlabel('$d_i$', fontsize=13)
plt.ylabel('$c(d_i)$', fontsize=13)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.legend(fontsize=11)

plt.xlim(-5, 5)

f.savefig("amse-function.pdf", bbox_inches='tight')
```

## Activation functions

```python
f = plt.figure()

x_start = -3
x_end = 3

x = np.linspace(x_start, x_end)
relu = np.linspace(x_start, x_end)
elu = np.linspace(x_start, x_end)

relu[relu<0] = 0
relu[relu>0] = x[x>0]

elu[elu<0] = np.exp(x[x<0]) - 1
elu[elu>0] = x[x>0]

plt.plot(x, 1 / (1 + np.exp(-x)), 'k', label=r"Sigmoid")
plt.plot(x, relu, 'k:', label=r"ReLU")
plt.plot(x, elu, 'k-.', label=r"ELU")
plt.xlabel('$x$', fontsize=13)
plt.ylabel('$f(x)$', fontsize=13)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.legend(fontsize=11);

plt.grid(True)
plt.xlim(-3, 3)
plt.ylim(-1, 2)

f.savefig("activation-functions.pdf", bbox_inches='tight')
```
